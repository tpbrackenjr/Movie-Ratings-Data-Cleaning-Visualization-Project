```r
library(tidyverse)
```

```r
df <- read_csv("C:/Users/tpbra/Downloads/movie_ratings_clean.csv")
```

```r
ggplot(df, aes(x = reorder(movie_title, -user_rating), y = user_rating)) +
  stat_summary(fun = mean, geom = "bar", fill = "steelblue") +
  labs(title = "Average Movie Rating", x = "Movie Title", y = "Average Rating") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
ggplot(df, aes(x = user_rating)) +
  geom_histogram(binwidth = 0.5, fill = "darkorange", color = "black") +
  labs(title = "Distribution of Movie Ratings", x = "Rating", y = "Count") +
  theme_minimal()
```

```r
ggplot(df, aes(y = user_rating)) +
  geom_boxplot(fill = "orange", color = "black") +
  labs(title = "Boxplot of Movie Ratings", y = "Rating") +
  theme_minimal()
```
